```python
import os

# 定义文件夹路径和新的文件名前缀
folder_path = "./background/"
new_name_prefix = "mask"

# 获取文件夹内所有图片的路径
image_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.png')]

# 遍历图片路径并逐一重命名
for i, p in enumerate(image_paths):
    # 构造新的文件名
    new_name = "{:03d}_{}.png".format(i, new_name_prefix)

    # 生成新的文件路径
    new_path = os.path.join(folder_path, new_name)

    # 重命名文件
    os.rename(p, new_path)

    # 打印提示信息
    print("{} -> {}".format(p, new_path))
```

```python
import os

# 定义文件夹路径和新的文件名前缀
folder_path = "./view/"
new_name_prefix = "mask"

# 获取文件夹内所有图片的路径
image_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.png')]

# 指定起始编号
start_number = 204

# 遍历图片路径并逐一重命名
for i, p in enumerate(image_paths):
    # 构造新的文件名
    new_name = "{:03d}_{}.png".format(i + start_number, new_name_prefix)

    # 生成新的文件路径
    new_path = os.path.join(folder_path, new_name)

    # 重命名文件
    os.rename(p, new_path)

    # 打印提示信息
    print("{} -> {}".format(p, new_path))
```

```python
import os
from PIL import Image
import shutil

# 定义原始文件夹路径和新的文件夹路径
folder_path = "./view/"
new_folder_path = "./"

# 创建新的目标文件夹
os.makedirs(new_folder_path, exist_ok=True)

# 遍历文件夹内所有图片并剪切
for f in os.listdir(folder_path):
    if f.endswith('.png'):
        # 打开图片，并获取图片尺寸
        image_path = os.path.join(folder_path, f)
        with Image.open(image_path) as im:
            width, height = im.size

            # 裁剪图片（去掉1/4大小的区域）
            box = (width // 4, height // 4, 3 * width // 4, 3 * height // 4)
            new_im = im.crop(box)

            # 生成新的文件路径，并保存裁剪后的图片
            new_image_path = os.path.join(new_folder_path, f)
            new_im.save(new_image_path)

# 删除原始文件夹及其内容
shutil.rmtree(folder_path)
```
