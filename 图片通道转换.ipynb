```python
import os
from PIL import Image
import torchvision.transforms as transforms
import torch

# 定义转换
transform = transforms.Compose([
    transforms.Resize((224, 224)),   # 调整尺寸
    transforms.ToTensor()            # 转为张量
])

# 指定文件夹路径
folder_path = r"D:\workspace\new_experiments\MMR-master\mvtec_anomaly_detection\bottle\train\good'

# 遍历文件夹中的所有图像文件
for file_name in os.listdir(folder_path):
    if file_name.endswith('.png'):
        # 读取图像
        img = Image.open(os.path.join(folder_path, file_name))
        
        # 将三通道图像转换为单通道图像
        img = img.convert('L')
        
        # 转换图像并添加批次维度
        img_tensor = transform(img).unsqueeze(0)
        
        # 打印图像张量的形状
        print(file_name, img_tensor.shape)
```

```python
import os
from PIL import Image
import torchvision.transforms as transforms

# 定义转换
transform = transforms.Compose([
    transforms.Resize((224, 224)),   # 调整尺寸
    transforms.Grayscale(),          # 转换为单通道图像
])

# 指定文件夹路径
folder_path = r"./ground_truth/deformation/"

# 遍历文件夹中的所有图像文件
for file_name in os.listdir(folder_path):
    if file_name.endswith('.png'):
        # 读取图像
        img = Image.open(os.path.join(folder_path, file_name))
        
        # 转换图像
        img_bw = transform(img)
        
        # 保存单通道图像到文件夹
        img_bw.save(os.path.join(folder_path, file_name))
```
